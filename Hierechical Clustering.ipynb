```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

# Llamamos a la dataframe
df = pd.read_csv("C:/Users/lucho/Desktop/Programacion/Python/Linkedin Data Science/base_datos_2008.csv", nrows = 1e4)
newdf = df[["AirTime","DepDelay"]].dropna()
```

```python
# Llamamos a el entrenamiento de agglomerativeClustering

clstr = AgglomerativeClustering(n_clusters = 5)
clstr.fit(newdf)

# distancias (euclidean / l1 / l2 / manhattan / cosine / precomputed)
# affinity = "l1",  linkage = "complete"
```

```python
plt.scatter(newdf["AirTime"], newdf["DepDelay"], c = clstr.fit_predict(newdf))
plt.show()
```
