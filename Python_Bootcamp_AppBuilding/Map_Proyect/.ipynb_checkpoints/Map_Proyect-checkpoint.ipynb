```python
import pandas as pd
import folium as fo
import numpy as np
```

```python
map = fo.Map()
```

```python
url = "https://raw.githubusercontent.com/lmoreno98/100DaysOfCode/main/Map_volcano/volcano.csv"
volcano = pd.read_csv(url)
volcano.tail()
```

```python
x = fo.FeatureGroup(name = "My Map")
```

```python
volcano_latitude = list(volcano["Latitude"])
volcano_longitude = list(volcano["Longitude"])
volcano_name = list(volcano["Name"])
final_list = [volcano_latitude,volcano_longitude,volcano_name]
```

```python
final = []
aux = []
for i in range(len(volcano_latitude)):
  aux = [volcano_latitude[i], volcano_longitude[i], volcano_name[i]]
  final.append(aux)
```

```python
for lat, lon, name, in final:
  x.add_child(fo.Marker(location=[lat,lon],
                        popup = name,
                        icon = fo.Icon(color = "red")))
```

```python
map.add_child(x)
```
