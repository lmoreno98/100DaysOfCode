# Etiquetas y legendas

```python
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/lucho/Desktop/Programacion/Python/Linkedin Data Science/base_datos_2008.csv", nrows = 100000)
data = np.unique(df.DayOfWeek, return_counts = True)
labs = ["Lun","Mar","Mie","Jue","Vie","Sab","Dom"]
data
```

```python
# Grafico de barras sin modificacion
plt.pie(x = data[1],
       labels = data[0],
       radius = 1.5,
       colors = ["Red","Green","Orange","Blue","Gray","Pink","Black"],
       startangle = 90)
plt.show()
```

```python
# Gradico de barras modificando todos los datos
plt.pie(x = data[1],
       #labels = labs,
       radius = 1,
       colors = sns.color_palette("hls",7),
        explode = (0,0,0,0.2,0,0,0.1),
        autopct = "%1.1f%%",
        startangle = 90)
plt.legend(loc = "upper right", labels = labs)
plt.show()
```

```python
# Creamos un grafico de barras con seaborn
plt = sns.barplot(x = labs, y = data[1])
plt.set(xlabel = "Dia de la semana", ylabel = "Numero de vuelos")
```

# Graficos para series temporales

```python
import datetime
import time
```

```python
#Creamos una variable que contenga solos los vuelos que tengan origen en PHX, HOU, IND
df = pd.read_csv("C:/Users/lucho/Desktop/Programacion/Python/Linkedin Data Science/base_datos_2008.csv")
df2 = df[df["Origin"].isin(["PHX","HOU","IND"])]
df = df.head(500000)
```

```python
# Lo que estamos haciendo es itterar por cada valor del df y que nos devuelva el aÃ±o mes y dia del aÃ±o
times = []
for i in np.arange(len(df)):
    times.append(datetime.datetime(year = 2008, month = df.loc[i,"Month"], day = df.loc[i,"DayofMonth"]))
```

```python
times[50000]
```

```python
df["Time"] = times
df.head()
```

```python
# Agrupamos por dia y calculculamos la media del ArrDelay y DepDelat
data = df.groupby(by = ["Time"], as_index = False)["DepDelay","ArrDelay"].mean()
data.head()
```

```python
sns.lineplot(x = data["Time"], y= data["DepDelay"])
```

```python
data = df.groupby(by=["Time"])["DepDelay","ArrDelay"].mean()
data.head()
```

```python
sns.lineplot(data = data)
```

```python
# Lo que estamos haciendo es itterar por cada valor del df y que nos devuelva el aÃ±o mes y dia del aÃ±o para el segunda df
times = []
for i in df2.index:
    times.append(datetime.datetime(year = 2008, month = df2.loc[i,"Month"], day = df2.loc[i,"DayofMonth"]))
df2["Time"] = times
```

```python
df2.head()
```

```python
sns.set(rc = {"figure.figsize": (15,10)})
sns.lineplot(x = "Time",y ="ArrDelay", hue = "Origin", data = df2)
```
