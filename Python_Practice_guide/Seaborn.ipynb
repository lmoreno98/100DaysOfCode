```python
# 1
iris = pd.read_html("https://github.com/lmoreno98/100DaysOfCode/blob/main/Python_Practice_guide/Datasets/Iris.csv")[0].drop(columns=["Unnamed: 0","Id"])
iris
```

```python
sns.countplot(iris["Species"],data=iris)
plt.title('Frequency of the three species in the Iris Dataset')
plt.ylabel('Frequency')
plt.ylim(0,80)
```

```python
# 2
sns.scatterplot(x= "SepalLengthCm", y = "SepalWidthCm", data= iris, hue="Species")
plt.title("Sepal lenght - Sepal Width")
```

```python
sns.scatterplot(x= "PetalLengthCm", y = "PetalWidthCm", data= iris, hue="Species")
plt.title("Petal lenght - Petal Width")
```

```python
# 3
iris
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4,1, figsize = (10, 10))
sns.histplot(x = 'SepalLengthCm', data = iris, ax = ax1)
ax1.set(title = 'Sepal Length histogram', ylabel = 'Frequency')

sns.histplot(x = 'SepalWidthCm', data = iris, ax = ax2)
ax2.set(title = 'Sepal Width histogram', ylabel = 'Frequency')

sns.histplot(x = 'PetalLengthCm', data = iris, ax = ax3)
ax3.set(title = 'Petal Length histogram', ylabel = 'Frequency')

sns.histplot(x = 'PetalWidthCm', data = iris, ax = ax4)
ax4.set(title = 'Petal Width histogram', ylabel = 'Frequency')

plt.tight_layout(pad = 2)
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4,1, figsize = (10, 10))
sns.boxplot(x = 'SepalLengthCm', data = iris, ax = ax1)
ax1.set(title = 'Sepal Length histogram', ylabel = 'Frequency')

sns.boxplot(x = 'SepalWidthCm', data = iris, ax = ax2)
ax2.set(title = 'Sepal Width histogram', ylabel = 'Frequency')

sns.boxplot(x = 'PetalLengthCm', data = iris, ax = ax3)
ax3.set(title = 'Petal Length histogram', ylabel = 'Frequency')

sns.boxplot(x = 'PetalWidthCm', data = iris, ax = ax4)
ax4.set(title = 'Petal Width histogram', ylabel = 'Frequency')

plt.tight_layout(pad = 2)
```

```python
corr = iris.corr()
sns.heatmap(corr, cmap="Blues", annot= True)
```

```python
sns.pairplot(iris)
```
