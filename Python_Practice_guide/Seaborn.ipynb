    Use the sns.joinplot() function to demonstrate the relationship between variables 'A' and 'B', showing their respective distributions. Repeat the same procedure for variables 'D' and 'F'.


```python
np.random.seed(20)
n = 2_000  #size
data = {
    'A': np.random.normal(loc = 100, scale = 1, size = n),
    'B': np.random.normal(loc = 100, scale = 2, size = n),
    'C': np.random.normal(loc = 100, scale = 5, size = n),
    'D': np.random.lognormal(mean = 2, sigma = 1.1, size = n),
    'E': np.random.lognormal(mean = 2, sigma = 1.2, size = n),
    'F': np.random.lognormal(mean = 2, sigma = 1.3, size = n),
    'G': np.random.randn(n),
    'H': np.random.binomial(n = 10, p = 0.8, size = n)
}

df = pd.DataFrame(data)
```

```python
# 16
sns.kdeplot(x="A", data = df)
sns.kdeplot(x="B", data = df)
sns.kdeplot(x="C", data = df)
plt.legend(["A","B","C"])
```

```python
#17
sns.kdeplot(x="D", data = df)
sns.kdeplot(x="E", data = df)
sns.kdeplot(x="F", data = df)
plt.legend(["D","E","F"])
```

```python
sns.jointplot(x = "A", y = "B", data = df)
```

```python
sns.jointplot(x = "D", y = "F", data = df)
```
