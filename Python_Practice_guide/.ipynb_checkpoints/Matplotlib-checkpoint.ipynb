## Exercise 4:
Use the DataFrame below to build a line graph, for the values in the 'Brand 1' column as a function of time.


    import matplotlib.pyplot as plt
    import pandas as pd
    import numpy as np
    n = 100
    columns = ['Brand '+str(number) for number in range(1,5)]
    date = pd.date_range(start = '2021-1-1', periods = n)
    data = np.random.randint(low = 1, high = [100, 330, 45, 77], size = (n, 4))
    df = pd.DataFrame(data,columns = columns, index = date)


Expected result:
<img src = https://img-c.udemycdn.com/redactor/raw/article_lecture/2021-01-13_23-31-31-fe8ebbb6462502f272795fbb491ce8ef.jpeg>

```python
n = 100
columns = ['Brand '+str(number) for number in range(1,5)]
date = pd.date_range(start = '2021-1-1', periods = n)
data = np.random.randint(low = 1, high = [100, 330, 45, 77], size = (n, 4))
df = pd.DataFrame(data,columns = columns, index = date)
df
```

```python
plt.figure(figsize=(12,5))
plt.plot(df["Brand 1"])
plt.title("Brand A")
plt.xlabel("Time")
plt.ylabel("Values")
```

## Exercise 9:
Use the data below to build the chart below. Change the font size to 15 on the axes and 20 on the title. Use the plt.legend() function to insert a legend on the chart, use the loc argument to set the position. Possibilities for setting the loc argument: best, upper right, upper left, lower left, lower right, right, center left, center right, lower center, upper center, center.


    import numpy as np
    import matplotlib.pyplot as plt
     
    x = np.arange(1,11)
    y1 = x**2
    y2 = x**3
    y3 = x**4


Expected result:
<img src = https://img-c.udemycdn.com/redactor/raw/article_lecture/2021-01-14_17-15-15-6420b0f1a325646c27f9dc091b935269.jpeg> 

```python
x = np.arange(1,11)
y1 = x**2
y2 = x**3
y3 = x**4
```

```python
plt.plot(y1,"o",y2,"^",y3,"bs")
plt.legend(["y1","y2","y3"])
plt.grid(True,color = "red", linestyle = "--", alpha = 0.8)
plt.xlabel("X", fontsize = 15)
plt.ylabel("Y", fontsize = 15)
plt.title("Values", fontsize = 20)
```

## Exercise 12:
Go back to the previous graph and add captions with the function plt.legend() (loc = 'upper left'), change the shadow argument to True and show the percentage of sales for each brand.

```python
sales = [1500.2, 1378, 671, 431, 700]
labels = ['Brand '+str(number) for number in range(1,len(sales)+1)]
plt.figure(figsize=(9,6))
plt.pie(x = sales, labels=labels, shadow= True, autopct='%1.0f%%')
plt.title("Pie Chart")
plt.legend(loc = "upper left",shadow =True)
plt.show()
```

Build the DataFrame below. Subsequently draw up the graph below. Change the limit of the y axis to 80-100, with the function plt.ylim(), and the size of the figure to (15.5) (use plt.figure (figsize = (15.5)).


    import numpy as np

    data = np.random.randint(low = 100, high = [250, 300, 500, 1000], size = (12, 4))
<img src = https://img-c.udemycdn.com/redactor/raw/article_lecture/2021-01-14_19-55-20-c697fae4c532bd64956eaaa86a94baa1.png>


Expected result:


<img src = https://img-c.udemycdn.com/redactor/raw/article_lecture/2021-01-14_19-56-54-3ae370e2bfb1b9b2f542da16eee128c3.jpeg>

```python
np.random.seed(7)
data = np.random.randint(low = 100, high = [250, 300, 500, 1000], size = (12, 4))
index = pd.date_range(start = '2021-1-1', periods = 12, freq = 'M').month_name()
ex13 = pd.DataFrame(data,index, columns=list("ABCD"))
ex13
```

```python
plt.figure(figsize=(15,5))
plt.plot(ex13["A"])
plt.plot(ex13["B"])
plt.plot(ex13["C"])
plt.plot(ex13["D"])
plt.grid(alpha = 0.5, linestyle = "--")
plt.legend(list("ABCD"))
plt.ylim(80, 1000)
```
