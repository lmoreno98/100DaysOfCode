```python
ex32["Final Avg"] =  ex32.drop(["Students"], axis = 1).mean(axis = 1)
```

```python
ex32
```

```python
def result(final_result):
    if final_result >= 7:
        return "Approved"
    else:
        return "Reproved"
ex32["Result"] = ex32["Final Avg"].apply(result)
```
