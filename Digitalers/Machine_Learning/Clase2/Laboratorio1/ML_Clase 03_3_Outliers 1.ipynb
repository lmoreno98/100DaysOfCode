```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set()
```

```python
np.random.seed(50)
n = 950 # para generar datos sin outliers ponemos n = 1000
datos_normales = np.random.normal(5,2, size = n)#centro, desviacion standard, tama√±o
datos_out = np.random.normal(12,1, size = 1000 - n)
```

```python
sns.distplot(datos_normales)
sns.distplot(datos_out)
```

```python
datos = np.concatenate((datos_normales, datos_out))
```

```python
sns.distplot(datos)
```

```python
sns.boxplot(datos)
```

```python
data = pd.DataFrame()
data['Datos'] = datos
data['Outlier'] = np.zeros(datos.size)
data.iloc[n:,1] = 1
```

```python
data = data.sample(frac = 1,random_state = 42)
data.reset_index(drop = True, inplace = True)
```
