# Visualization of quantitative data
* Numerical data

```python
import numpy as np
```

### Histograms

```python
sns.histplot(airbnb["price"])
plt.xlabel('Price(in US Sollars)')
```

```python
sns.histplot(airbnb["price"], bins = np.arange(0,1100,40))
plt.xlabel('Price(in US Sollars)')
```

### Scatterplot

```python
sns.scatterplot(x = 'price', y = 'number_of_reviews', data = airbnb)
plt.xlabel("Price")
plt.ylabel("Number of reviews")
```

```python
sns.scatterplot(x = 'price', y = 'number_of_reviews', data = airbnb)
plt.xlabel("Price")
plt.ylabel("Number of reviews")
plt.xlim(0,1100)
```

```python
sns.scatterplot(x = 'price', y = 'number_of_reviews', data = airbnb,size= 5)
plt.xlabel("Price")
plt.ylabel("Number of reviews")
plt.xlim(0,1100)
```
