# Missing values

```python
df["Time_taken"].fillna(df["Time_taken"].mean(), inplace=True)
```

```python
df.info()
```

# Dummy variables

```python
df = pd.get_dummies(df, drop_first= True)
```

```python
df.head()
```

# X-y split

```python
X = df.iloc[:,df.columns != "Collection"]
X.head()
```

```python
y = df["Collection"]
y.head()
```

# Train Test slpit

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Standarizing Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler().fit(X_train)
```

```python
X_train_std = sc.transform(X_train)
```

```python
X_test_std = sc.transform(X_test)
```

```python
X_test_std
```

# Training model

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel="linear", C = 1000)
```

```python
svr.fit(X_train_std, y_train)
```

# Predicted values

```python
y_test_pred = svr.predict(X_test_std)
y_train_pred = svr.predict(X_train_std)
```

```python
y_test_pred
```

# Model performance

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
mean_squared_error(y_test,y_test_pred)
```

```python
r2_score(y_test, y_test_pred)
```

```python
r2_score(y_train, y_train_pred)
```

# Grid performance

```python
from sklearn.model_selection import GridSearchCV
```

```python
svr = SVR()
```

```python
params_grid = {"kernel":["linear","rbf"], 
            "C": [250,500,1000,2000,5000]}
```

```python
grid_search = GridSearchCV(svr,params_grid,
                          n_jobs=-1,cv=5,scoring="r2")
```

```python
grid_search.fit(X_train_std, y_train)
```

```python
grid_search.best_params_
```

```python
cv_svr = grid_search.best_estimator_
```

```python
r2_score(y_test, cv_svr.predict(X_test_std))
```
