# Missing values

```python
df["Time_taken"].fillna(df["Time_taken"].mean(), inplace=True)
```

```python
df.info()
```

# Dummy variables

```python
df.head()
```

```python
df = pd.get_dummies(df, columns= ["3D_available", "Genre"], drop_first = True)
```

```python
df.head()
```

# X-y split

```python
X = df.iloc[:,df.columns != "Collection"]
X.head()
```

```python
print(f"Type of X : {type(X)}")
print(f"Shape of X : {X.shape}")
```

```python
y = df["Collection"]
y.head()
```

```python
print(f"Type of y : {type(y)}")
print(f"Shape of y : {y.shape}")
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
print(f"""X_train shape: {X_train.shape}
X_test shape: {X_test.shape}
y_train shape: {y_train.shape}
y_test shape: {y_test.shape}""")
```

```python
X_train.head()
```

# Training Regression Tree

```python
from sklearn import tree
regtree = tree.DecisionTreeRegressor(max_depth = 3)
```

```python
regtree.fit(X_train, y_train)
```

# Predict values

```python
y_train_pred = regtree.predict(X_train)
y_test_pred = regtree.predict(X_test)
```

```python
y_test_pred
```

# Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score
```

Mean squared error is for calculate performance between diferent models

```python
mean_squared_error(y_test, y_test_pred)
```

```python
r2_score(y_train,y_train_pred)
```

```python
r2_score(y_test,y_test_pred)
```

# Plotting Decision tree

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(regtree)
```

# Prunnig tree

### Maximum number of levels in tree

```python
regtree1 = tree.DecisionTreeRegressor(max_depth=3)
regtree1.fit(X_train, y_train)
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(regtree1,feature_names= X_train.columns, filled= True )
```

### Minimum observation at internal node

```python
regtree2 = tree.DecisionTreeRegressor(min_samples_split=40)
regtree2.fit(X_train, y_train)
fig = plt.figure(figsize=(40,35))
_ = tree.plot_tree(regtree2,feature_names= X_train.columns, filled= True)
```

```python
regtree3 = tree.DecisionTreeRegressor(min_samples_leaf= 25)
regtree3.fit(X_train, y_train)
fig = plt.figure(figsize=(40,35))
_ = tree.plot_tree(regtree3,feature_names= X_train.columns, filled= True)
```
