# Walk Foward Validation

```python
df.head()
```

def parser(x): return datetime, strptime('190'+x, '%Y-%m')

```python
train_size = int(df.shape[0]*0.7)
```

```python
train, test = df.Sales[0:train_size], df.Sales[train_size:]
```

```python
train.shape
```

```python
test.shape
```

```python
data = train
predict = []
for t in test:
    model = ARIMA(data, order=(4,1,0))
    model_fit = model.fit()
    y = model_fit.forecast()
    print(y[0][0])
    predict.append(y[0][0])
    data = np.append(data,t)
    data = pd.Series(data)
```

```python
predict
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test.values, predict)
mse
```
