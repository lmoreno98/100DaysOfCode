```python
import pandas as pd
import os
import random
import matplotlib.pyplot as plt
```

```python
def upload_data():
    dirs = os.listdir()
    choice = input("Do you want to see the files that you have Y or N: ").upper()
    if choice == "Y":
        for dir in dirs:
            if(dir.endswith(".csv") or dir.endswith(".json") or dir.endswith(".txt")):
                print(dir)
    elif choice == "N":
        pass
    else:
        print("Try again please select between Y or N")
        return False
    path = input("Insert the name of the file: ")
    
    if path.endswith(".csv"):
        dataframe = pd.read_csv(path)
    elif path.endswith(".json"):
        dataframe = pd.read_json(path)
    elif path.endswith(".txt"):
        dataframe = pd.read_csv(path)
    return dataframe
```

```python
def random_data():
    df = pd.DataFrame()
    lista=[]
    for i in range(100):    
        lista.append(random.random())
    df["A"] = lista
    lista=[]
    for i in range(100):    
        lista.append(random.random())
    df["B"] = lista
    lista=[]
    for i in range(100):    
        lista.append(random.random())
    df["C"] = lista
    lista=[]
    for i in range(100):    
        lista.append(random.random())
    df["D"] = lista
    lista=[]
    for i in range(100):    
        lista.append(random.random())
    df["E"] = lista
    return df
```

```python
def create_df():
    df = pd.DataFrame()
    lista = []
    name = []
    print("Enter the values for columns name")
    for i in range(5):
        name.append(input(""))
    print("Enter the element of the first row")
    for i in range(5):
        lista.append(input(""))
    df[name[0]] = lista
    lista = []
    print("Enter the element of the second row")
    for i in range(5):
        lista.append(input(""))
    df[name[1]] = lista
    lista = []
    print("Enter the element of the third row")
    for i in range(5):
        lista.append(input(""))
    df[name[2]] = lista
    lista = []
    print("Enter the element of the fourth row")
    for i in range(5):
        lista.append(input(""))
    df[name[3]] = lista
    return df
```

```python
def createdata(data):
    if data == 1:
        return random_data()
    elif data == 2:
        return create_df()
    elif data == 3:
        return upload_data()
```
