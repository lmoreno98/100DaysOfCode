# GroupBy

```python
df = pd.read_csv("C://Users/lucho/Desktop/Programacion/Python/Linkedin Data Science/base_datos_2008.csv", nrows= 100000)
```

```python
# Nos da el retraso maximo que hay en cada dia de la semana 
df.groupby(by = "DayOfWeek")["ArrDelay"].max()
```

```python
# Nos da la media de retraso que hay en cada dia de la semana 
df.groupby(by = "DayOfWeek")["ArrDelay"].mean()
```

```python
# Nos da el retraso minimo que hay en cada dia de la semana 
df.groupby(by = "DayOfWeek")["ArrDelay"].min()
```

```python
# Nos da la cantidad de vuelos que hay por dias de semana
df.groupby(by = "DayOfWeek")["ArrDelay"].count()
```

```python
# Nos da el resumen estadistico de todo los que pasa en cada dia de la semana 
df.groupby(by = "DayOfWeek")["ArrDelay"].describe()
```

```python
# Nos da la media de "ArrDelay" y "DepDelay" y nos lo muestra en la misma tabla por dia de la semana
df.groupby(by = "DayOfWeek")["ArrDelay","DepDelay"].mean()
```

```python
# Rango de valores para cada dia de la semana
df.groupby(by = "DayOfWeek")["ArrDelay"].max() - df.groupby(by = "DayOfWeek")["ArrDelay"].min()
```

```python
dfPHXHOU = df[df.Origin.isin(["PHX","HOU"])]
```

```python
#Nos muestra la media de retraso por dia de la semana y por destino de viaje 
dfPHXHOU.groupby(by = ["DayOfWeek", "Origin"])["ArrDelay"].mean()
```

```python
# Guardamos el groupby
mygroupby = dfPHXHOU.groupby(by = ["DayOfWeek", "Origin"])["ArrDelay"]
```

#                               Datos duplicados y datos nulls

```python
# 1e6 = 1,000,000
df = pd.read_csv("C://Users/lucho/Desktop/Programacion/Python/Linkedin Data Science/base_datos_2008.csv", nrows= 1e6)
```

## Datos duplicados

```python
# Creamos el dataframe duplicado por el mismo para tener la misma cantidad de valores duplicados
df_duplicated = df.append(df)
```

```python
# Desordenamos los datos
df_duplicated = df_duplicated.sample(frac = 1)
df_duplicated
```

```python
# Eliminamos los duplicados
dfclean = df_duplicated.drop_duplicates()
dfclean
```

```python
# Como habia valores duplicados en el dataframe original nos da  false
len(dfclean) == len(df)
```

## Drop suplicates

```python
# Guardamos una dataframe que guarda los valores de un solo dia del mes
dfclean.drop_duplicates(subset = "DayofMonth")
```

```python
# Nos da vacio porque no hay ningun valor en nuestra fila que no tenga ningun dato faltante
df.dropna()
```

```python
# Thresh nos deja elegir cuantas columnas queremos con valores null en este caso queremos que 25% tengan null
df.dropna(thresh = 25)
```
