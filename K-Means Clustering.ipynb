```python
import pandas as pd
import numpy as np 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Llamamos a la dataframe
df = pd.read_csv("C:/Users/lucho/Desktop/Programacion/Python/Linkedin Data Science/base_datos_2008.csv", nrows = 1e5)
newdf = df[["AirTime","DepDelay"]].dropna()
```

```python
# Llamamos a el entrenamiento de kmeans
kmeans = KMeans(n_clusters = 4, random_state = 0).fit(newdf)

# muestro el output del array
kmeans.labels_
```

```python
# Muestra como desta distribuido todos los grupos
np.unique(kmeans.labels_, return_counts=True)
```

```python
# Grafico para ver de forma mas comoda como estan distribuido los puntos
plt.scatter(newdf["AirTime"], newdf["DepDelay"], c = kmeans.labels_)
# Como vemos en el grafico hay mas dispercio del grafico violeta y amarillo que de los otros dos
plt.show
```

```python
# Muestro los centros de los clusters
kmeans.cluster_centers_
```

```python
# Cargamos una nueva base de datos para predecir
alldf =  pd.read_csv("C:/Users/lucho/Desktop/Programacion/Python/Linkedin Data Science/base_datos_2008.csv", nrows = 1e6)[["AirTime","DepDelay"]].dropna()
```

```python
# Clasificacion de los primeros 50
kmeans.predict(alldf)[0:50]
```
